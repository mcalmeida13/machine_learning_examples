```python
df = pd.read_csv('site_data.csv',names=['last_page_id','next_page_id'])
df.head()
```

```python
#COLLECT COUNTS
grouped_df = df.groupby(df.columns.tolist()).size().reset_index().rename(columns={0:'records'})
# grouped_df2 = df.groupby().size().reset_index().rename(columns={0:'records'})
grouped_df
# grouped_df2
```

```python
grouped_df['transition_proba'] = grouped_df['records']/grouped_df.groupby('last_page_id')['records'].transform('sum')
```
